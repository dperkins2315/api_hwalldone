```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

# Import API key
api_key= input ('enter your own api key for open weather ')

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy
from time import strftime,localtime

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#use for notes:
#url="http://api.openweathermap.org/data/2.5/weather?appid=" + api_key + "&units=Imperial&q="
# response = requests.get(url + "london").json()
#print(json.dumps(response,indent=4))
```

```python
# # Save config information
# url = "http://api.openweathermap.org/data/2.5/weather?"
# city = "London"

# # Build query URL
# query_url = url + "appid=" + api_key + "&q=" + city

url="http://api.openweathermap.org/data/2.5/weather?appid=" + api_key + "&units=Imperial&q="

#hold retreived data
orig_data= {'city':[],
            'cloudiness':[],
            'country':[],
            'date':[],
            'humidity':[],
            'lat':[],
            'lng':[],
            'max temp':[],
            'wind speed':[]}
            
#count log record #
counter= 1

#data retreval place
print("Data Log")
print('-'*35)
      
#loop
for i in cities:
    response= requests.get(url + i).json()
    if response["cod"] != '404':
        print("Getting data for city: {}, record {}".format(i,counter))
        orig_data["city"].append(response["name"])
        orig_data["cloudiness"].append(response["clouds"]["all"])
        orig_data["country"].append(response["sys"]["country"])
        orig_data["date"].append(response["dt"])
        orig_data["humidity"].append(response["main"]["humidity"])
        orig_data["lat"].append(response["coord"]["lat"])
        orig_data['lng'].append(response['coord']['lon'])
        orig_data["max temp"].append(response["main"]["temp_max"])
        orig_data["wind speed"].append(response["wind"]["speed"])
        counter += 1
        #sleep 
        time.sleep(1.05)    
    else:
        print("city {} not found. Skip it".format(i))
        time.sleep(1.05)
print("End of Log")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
file= 'weatherdata.csv'
weather_data= pd.DataFrame(orig_data)
#weather_data.head()
weather_data.to_csv(file,index=False)



            
```

### Plotting the Data

#### Latitude vs. Temperature Plot

```python
plt.scatter(weather_data['lat'], weather_data['max temp'], alpha= 0.6, c="green", edgecolor="black")
plt.grid()
plt.xlabel("Latitude")
plt.ylabel("Max Temp (Fahrenheit)")
plt.title("Latitude vs. Temperature ({})".format(strftime("%m/%d/%Y",localtime())))
plt.savefig("latvstemp.png")
```

#### Latitude vs. Humidity Plot

```python
plt.scatter(weather_data['lat'], weather_data['humidity'], alpha= 0.6, c="grey", edgecolor="black")
plt.grid()
plt.xlabel("Latitude")
plt.ylabel("Humidity")
plt.title("Latitude vs. Humidity({})".format(strftime("%m/%d/%Y",localtime())))
plt.savefig("latvshumid.png")
```

#### Latitude vs. Cloudiness Plot

```python
plt.scatter(weather_data['lat'], weather_data['cloudiness'], alpha= 0.6, c="blue", edgecolor="black")
plt.grid()
plt.xlabel("Latitude")
plt.ylabel("Cloudiness")
plt.title("Latitude vs. CLoudiness ({})".format(strftime("%m/%d/%Y",localtime())))
plt.savefig("latvsclouds.png")
```

#### Latitude vs. Wind Speed Plot

```python
plt.scatter(weather_data['lat'], weather_data['wind speed'], alpha= 0.6, c="purple", edgecolor="black")
plt.grid()
plt.xlabel("Latitude")
plt.ylabel("Wind Speed")
plt.title("Latitude vs. Wind Speed ({})".format(strftime("%m/%d/%Y",localtime())))
plt.savefig("latvswind.png")
```
